>### bit.ly&1.usa.gov

```python
path = 'ch02/usagov_bitly_data2012-03-16-1331923249.txt'
open(path).readline()
```

```python
import json
records = [json.loads(line) for line in open(path)]
records[0]
```

```python
from pandas import DataFrame,Series
frame = DataFrame(records)
frame.head()
```

```python
# cframe = frame['a'].dropna()
cframe = frame[frame.a.notnull()]
```

```python
import numpy as np

o_s = np.where(cframe.a.str.contains('Windows'),'Windows','Not Windows')
o_s[:10]
```

```python
by_tz_os = cframe.groupby(['tz',o_s]).size().unstack().fillna(0)
by_tz_os[:10]
```

```python
indexer = by_tz_os.sum(1).argsort()
indexer
```

```python
count_subset = by_tz_os.take(indexer)[-10:]
count_subset
```

```python
%pylab inline

normal_subset = count_subset.div(count_subset.sum(1),axis=0)
normal_subset.plot(kind='barh',stacked=True,figsize=(10,6))
```

>### MovieLens 1M数据集

```python
import pandas as pd
import os
encoding = 'latin1'

upath = os.path.expanduser('ch02/ml-1m/users.dat')
rpath = os.path.expanduser('ch02/ml-1m/ratings.dat')
mpath = os.path.expanduser('ch02/ml-1m/movies.dat')

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
mnames = ['movie_id', 'title', 'genres']

users = pd.read_csv(upath, sep='::', header=None, names=unames, encoding=encoding,engine='python')
ratings = pd.read_csv(rpath, sep='::', header=None, names=rnames, encoding=encoding,engine='python')
movies = pd.read_csv(mpath, sep='::', header=None, names=mnames, encoding=encoding,engine='python')
```

```python
users.head()
```

```python
ratings.head()
```

```python
movies.head()
```

```python
data = pd.merge(pd.merge(users,ratings),movies)
data.head()
```

```python
mean_ratings = data.pivot_table('rating',index='title',columns='gender',aggfunc=mean)
mean_ratings.head()
```

```python
#.agg([('count',size)])
ratings_500 = data.groupby('title').size()
ratings_500[:10]
```

```python
active_titles = ratings_500.index[ratings_500>=500]
active_titles
```

```python
#出错mean_ratings = mean_ratings[active_titles]
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings.head()
```

```python
mean_ratings.sort_values('F',ascending=False).head(10)
```

```python
mean_ratings['diff'] = mean_ratings['F']-mean_ratings['M']

sort_by_diff = mean_ratings.sort_values('diff',ascending=False)
sort_by_diff[:10]
```
